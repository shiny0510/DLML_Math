2. 통계학(statistics): 불확실한 현상을 대상으로 자료를 수집하고 정리해 이 자료가 수집된 대상에 대한 적절한 모형을 설정하고 추정(estimation), 검정(testing), 예측(forecasting)을 하는 학문이다.

8. 자료의 흩어진 정도를 측정하는 통계량

 * 표준편차(standard deviation): $S=\sqrt{\frac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2}$

7. 결합확률분포(Joint Probability distribution):

- 이산형 확률변수 (discrete random variable): 확률변수 X의 가능한 값이 $x_1, ... , x_n$ 이고 확률변수 Y의 가능한 값이 $y_1, ..., y_m$이라면, $(X,Y)$의 결합확률 분포는 다음과 같이 정의한다. 

<br>
    $$Pr(X=x_i,Y=y_j)= P_{ij}, i=1,2,...,n, j=1,2,...,m$$
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;여기에서 $Pr(X=x_i,Y=y_j)= P_{ij}$를 결합확률질량함수(joint probability mass function)라고 정의한다.

<br>

- 연속형 확률변수(continuous random variable) : 

  두 확률변수 X,Y의 결합확률밀도함수(joint probability density function)를 f(x,y)라 정의할때, 모든 (x,y)에 대해 $f(x,y)\geq0$ 이고,
<br>
$$\int_{-\infty}^\infty\int_{-\infty}^\infty f(x,y)dxdy=1$$
<br>
이다.

<br>

8. 주변확률분포(Marginal Probability Distribution):

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1) 이산형 확률변수: 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;결합분포가

<br>
    $$Pr(X=x_i,Y=y_j)= P_{ij}, i=1,2,...,n, j=1,2,...,m$$
<br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;이라 할 때, X와 Y의 주변확률분포는 각각 다음과 같이 구한다.
<br>
  - X의 주변확률분포:
<br>
    $$P(X=x_i)= \sum_{j=1}^m P_{ij}, (i=1,2,...,n)$$
<br>
  - Y의 주변확률분포:
<br>
    $$P(Y=y_j)= \sum_{i=1}^n P_{ij}, (j=1,2,...,n)$$
<br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2) 연속형 확률분포:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;두 확률변수 X,Y의 결합확률밀도함수를 $f(x,y)$라 할 때, X와 Y의 주변확률분포는 각각 다음과 같이 정의된다.


- X의 주변확률분포:
<br>
$$f(x)=\int_{-\infty}^\infty f(x,y)dy$$
<br>

- Y의 주변확률분포:
<br>
$$f(y)=\int_{-\infty}^\infty f(x,y)dx$$
<br>

 연속형 확률분포에서는 주변확률분포함수를 주변확률밀도함수(mariginal probability density function)라 부른다.

11. 이산형 확률변수: 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1) 이산형 확률변수:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;이산형 확률변수 X의 가능한 값이 $(x_1,x_2, ... , x_n)$이며,

$$P(X=x_i)= P_{i}, i=1,2,...,n$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;일때, X의 기댓값 $E(X)$는

$$E(X)= \sum_{i=1}^n x_i P_i$$

<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;2) 연속형 확률변수:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;연속형 확률변수 X의 확률밀도함수가 $f(x)$라면, X의 $E(X)$는

$$E(X)=\int_{-\infty}^{\infty}x \cdot f(x) dx$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;이다.

-  $E(aX+b)=aE(X)+b$
-  $E(aX+bY)=aE(X)+bE(Y)$

13. 분산

  확률변수 X의 분산(variance)은 $E(X)=\mu$라 할 때 X와 $\mu$의 편차의 제곱, 즉 $(X-\mu)^2$의 기댓값으로 $Var(X)$ 또는 $\sigma^2_x$으로 표현되며,

$$\sigma^2_x = E(X-\mu)^2$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;으로 구한다. 여기에서 확률변수 X를 나타낼 필요가 없는 경우에는 $\sigma^2_x$에서 X를 없애고 $\sigma^2$으로 표현한다. 분산의 양의 제곱근을 표준편차(standard deviation)라 하며, $\sigma$로 표현한다. 즉, 

$$\sigma=\sqrt{\sigma^2}$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;이다.


&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\sigma^2$은 X가 이산형 확률변수일 때 X의 가능한 값이 $(x_1,x_2,..,x_n)$이고 $P(X=x_i)=P_i$라면,

$$\sigma^2=E(X-\mu)^2= \sum_{i=1}^n (x_i-\mu)^2 P_i$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 이고, X가 연속형 확률변수이고 X의 확률밀도함수가 $f(X)$이면,

$$\sigma^2=E(X-\mu)^2=\int (x-\mu)^2 \cdot f(x) dx$$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;1) 분산의 계산공식 

$$\sigma^2 = E(X^2) - (E(X))^2 = E(X^2)-\mu^2$$

14. 공분산

두 변수 사이의 관계의 밀접도를 측정하는 상관계수를 구하는 과정에서 계산되는 경우가 많다.

두 확률변수 X와 Y의 공분산(covariance)은 $Cov(X,Y)$ 또는 $\sigma_{XY}$로 표현하며, 다음과 같이 계산한다.

$$\sigma = E[(X-\mu_X)(Y-\mu_Y)] = E(XY) - \mu_X\cdot\mu_Y$$
